# Analysis Performed

The following analyses were performed in this project:

# 1.Flight Status Distribution

Analyzed the number of flights that were On Time, Delayed, and Cancelled.

Visualized using a bar chart.

```python
#flight status
```

```python
import matplotlib.pyplot as plt
df["Flight Status"].value_counts().plot(kind="bar")
plt.title("Flight Status Distribution")
plt.xlabel("Flight Status")
plt.ylabel("Number of Flights")
plt.show()
```

# 2.Top Airports by Passenger Count

Identified the top 10 airports with the highest number of passengers.

Visualized using a bar chart.


```python
df["Airport Name"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Airports by Passenger Count")
plt.xlabel("Airport Name")
plt.ylabel("Number of Passengers")
plt.show()
#passenger count on from different airports
```

# 3.Top Countries by Passenger Traffic
