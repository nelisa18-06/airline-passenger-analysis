# Airline Passenger and Flight Operations Analysis

# - About the Dataset

This project uses an airline passenger dataset obtained from Kaggle.
The dataset contains information related to airline passengers, airports, flight status, and geographical details such as countries and continents.

The dataset includes attributes like passenger age, nationality, airport name, flight status, and departure details. It does not contain ticket pricing or revenue information and is mainly focused on passenger movement and flight operations

# Objective of the Project

The main objective of this project is to analyze airline passenger data to understand:

Flight operation performance

Passenger distribution across airports, countries, and continents

Basic passenger demographics

This project is designed as a beginner-level Data Analyst project to practice data validation, exploratory data analysis, and visualization using Python.

# Tools & Libraries Used

Python

Pandas (for data handling and analysis)

Matplotlib (for data visualization)

Jupyter Notebook
